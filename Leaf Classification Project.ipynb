```python
#Before outliers treatment
for i in train_df[train_df.columns[2:10]]:
    sns.boxplot(train_df[i])
    plot.show()
```

```python
for i in train_df.var().sort_values(ascending=False).index[2:50]:
    x=np.array(train_df[i])
    qr1=np.quantile(x,0.25)
    qr3=np.quantile(x,0.75)
    Iqr=qr3-qr1
    utv=qr3+(1.5*Iqr)
    ltv=qr1-(1.5*Iqr)
    y=[]
    for j in x:
        if j<ltv or j>utv:
            y.append(np.median(x))
        else:
            y.append(j)
    train_df[i]=y
            
```
